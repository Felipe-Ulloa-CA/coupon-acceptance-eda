# Will the Customer Accept the Coupon?

Exploratory Data Analysis using Pandas and Matplotlib.

```
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = '../data/coupons.csv'
TARGET_COL = 'y'

df = pd.read_csv(DATA_PATH)
df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]
df.head()
```

## Acceptance Rate by Coupon Type

```
coupon_col = [c for c in df.columns if 'coupon' in c][0]
acc = df.groupby(coupon_col)[TARGET_COL].mean().sort_values(ascending=False)
acc.plot(kind='bar')
plt.title('Acceptance Rate by Coupon Type')
plt.ylabel('Acceptance Rate')
plt.tight_layout()
plt.show()
acc
```

## Conclusions
- Coupon type significantly affects acceptance.
- Contextual variables (time, passenger, destination) are likely relevant.
- Future work: train a logistic regression model.
